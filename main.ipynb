```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/city_day.csv")

df.head()
```

```python
# AQI missing value since AQI is target attribute
target_missing_values = df["AQI"].isna().sum()
target_missing_values
```

```python
df_new = df.dropna(subset=['AQI'], ignore_index=True)
```

```python
df_new.describe()
```

```python
df_new.describe(include="object")
```

```python
df_new = df_new.iloc[:,2:15]
df_new.corr()
```

```python
df_new.info()
```

```python
from sklearn.impute import SimpleImputer
imp_median = SimpleImputer(missing_values=np.nan, strategy="median")
idf = pd.DataFrame(imp_median.fit_transform(df_new))
idf.columns=df_new.columns
idf.index=df_new.index
```

```python
idf.info()
```

```python
from sklearn.model_selection import train_test_split

X = idf.columns[:-1]
y = idf.columns[-1:]
X = idf[X]
y = idf[y]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
idf.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = regressor.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("SVR:")
print(f"RMSE = {rmse}, R2 = {r2}")
```

```python
import xgboost as xgb

model = xgb.XGBRegressor(objective ='reg:squarederror', random_state=42)  
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("XGBoost:")
print(f"RMSE = {rmse}, R2 = {r2}")
```
